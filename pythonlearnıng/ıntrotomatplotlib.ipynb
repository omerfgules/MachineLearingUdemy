```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

 %matplotlib inline

```python
age_list=[10,20,30,40,50,60,75]
```

```python
weight_list=[20,60,80,90,67,78,95] 
```

```python
plt.plot(age_list,weight_list,"r")#length must be same

#plt.savefig("test.png")##önemli
plt.title("Age vs weight")
plt.show()
```

```python
np_age_list=np.array(age_list)
```

```python
np_weight_list=np.array(weight_list)
```

```python
plt.plot(np_age_list,np_weight_list,"g")
plt.title("Weight vs Age")
plt.show()
```

```python
numpy_arr1=np.linspace(0,10,20)
```

```python
numpy_arr1
```

```python
numpy_arr2=numpy_arr1**3
```

```python
numpy_arr2
```

```python
plt.plot(numpy_arr1,numpy_arr2,"b--")
plt.show()
```

```python
plt.subplot(1,2,1)
plt.plot(numpy_arr1,numpy_arr2,"b--")
plt.subplot(1,2,2)#1 row 2 columns and creating second graph 
plt.plot(numpy_arr2,numpy_arr1,"g--")
plt.show()
```

```python
my_figure = plt.figure()
figure_axes=my_figure.add_axes([0.1,0.1,0.5,0.6])
figure_axes.plot
figure_axes.set_xlabel("X Axis")
figure_axes.set_ylabel("Y Axis ")
figure_axes.set_title("Graph title ")
plt.show()
```

```python
new_fig=plt.figure(dpi=1000)#dpi çözünürlük verisi veriyor
new_axes=new_fig.add_axes([0.1,0.1,0.9,0.9])
new_axes.plot(numpy_arr1,numpy_arr1**2,label="numpy arr **2")
new_axes.plot(numpy_arr1,numpy_arr1**3,label="numpy arr **3")
new_axes.legend(loc=3) #information
plt.show()
```
