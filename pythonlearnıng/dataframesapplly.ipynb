```python
def salary_category(salary):
    if salary<50000:
        return "Low"
    elif 80000>salary>=50000:
        return "Medium"
    else:
        return "high"
```

```python
df["Salary"].apply(salary_category)
```

```python
df["Salary_category"]=df["Salary"].apply(salary_category)
```

```python
df
```

```python
def tooexperienced(experience):
    if experience >=10:
        return 1
    else:
        return 0
```

```python
df["Experience"].apply(tooexperienced)
```

```python
df["New_perf_Score"]=(df["Experience"].apply(tooexperienced))+df["Performance_Score"]
```

```python
df
```

```python
def adjust_new(row):
    if row["Experience"]>10:
        return int(row["Performance_Score"]+1)
    else: 
        return row["Performance_Score"]
```

```python
df["Adjusted_score"]=df.apply(adjust_new,axis=1)
```

```python
df["Formatted_Name"]=df["Name"].apply(lambda x : x.replace("_"," "))
```
