```python
data =np.array([10,12,13,15,18,25,100,105])
```

```python
mean
```

```python
std=np.std(data)
```

```python
std
```

```python
z_scores=(data-mean)/std
```

```python
z_scores #outlier olan verilerin z score yuksek çıktı 
```

```python
z_scores[z_scores>1]
```
