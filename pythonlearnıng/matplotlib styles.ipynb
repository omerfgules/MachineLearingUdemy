```python
import numpy as np

import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data=np.linspace(0,10,20)
```

```python
data
```

```python
data2=data**2
```

```python
plt.subplots()
```

```python
my_fig,my_axes=plt.subplots()
my_axes.plot(data,data2, color="black",alpha=0.3)
my_axes.plot(data2,data, color="lightgreen")
plt.show()
```

```python
(new_fig,new_axes)=plt.subplots()
new_axes.plot(data,data+4,color="red",linestyle="-")
new_axes.scatter(data,data+1,color="red",linestyle="-")
new_axes.plot(data,data+2,color="red",linestyle="-")
new_axes.plot(data,data+5,color="red",linestyle="-",marker="o",markersize=4,markerfacecolor="yellow")
plt.show()
```

```python
plt.scatter(data,data2,color="black")
plt.show()
```
