```python
import numpy as np
import seaborn as sns
import pandas as np
import matplotlib.pyplot as plt
```

```python
#boxplot
```

```python
data=np.array([5,7,9,10,20,22,23,25,30,35,37,40,50,55,60,100])
plt.figure(figsize=(6,5),facecolor="Red")
sns.boxplot(y=data)
plt.title("Box plot")
plt.ylabel("data Value")
plt.grid(True)
plt.show()
```
